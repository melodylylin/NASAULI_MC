# SE(2) Theory

We consider the Lie Group SE(2):

$$\chi = \begin{bmatrix}
    R & p \\
    & 1                   
\end{bmatrix}$$
where $R \in SO(2)$, $p \in \mathbb{R}^2$

This is the Special Euclidean group of 2 dimensions. Special indicates that all matrices have determinant 1. The Special Euclidean group is the semi direct product of $SO(2)$ and $\mathbb{R}^2$.

The corresponding Lie Algebra se(2) is:

$$x = \begin{bmatrix}
    [\omega] & v \\
    & 0                   
\end{bmatrix}$$
where $\omega = -\omega^T$, $v \in \mathbb{R}^2$

Consider the dynamcis given in the Lie Group (note that $\chi$ is the system dynamics, and $\bar{\chi}$ is the reference dynamics to track:

$$
\dot{\chi} = \chi[v]\\
\dot{\bar{\chi}} = \bar{\chi}[\bar{v}]\\
$$

Note that the $[\cdot]$ indicates the Lie wedge operator that maps from $R^3$ to the the Lie algebra se(2).

$[\cdot] : R^3 \mapsto se(2), \begin{bmatrix}x\\ y\\ \theta \end{bmatrix} \to \begin{bmatrix} 0 & -\theta & x \\ \theta& 0 & y \\ 0& 0 & 0\end{bmatrix}$

Now define the error $\eta$.

\begin{align}
\eta &= \chi^{-1}\bar{\chi}\\
\end{align}

Using the product rule.

\begin{align}
\dot{\eta} &= -\chi^{-1}\dot{\chi}\chi{-1}\bar{\chi} + \eta^{-1}\dot{\bar{\chi}}\\
\dot{\eta} &= \eta[\bar{v}] -[v]\eta \\
\end{align}

The dynamics in the Lie Algebra are given exactly by the derivative of the exponential map. Here $1$ denotes the identity element of the group, for Matrix Lie groups, the identity matrix.

$$
\dfrac{d}{dt} exp([\zeta(t)]) = exp([\zeta(t)]) \dfrac{1 - exp([-ad_\zeta])}{ad_\zeta} \dfrac{d}{dt} \zeta(t)
$$

Let $\eta(t) = exp([\zeta(t)])$, we will drop the $(t)$ for these functions of time for brevity.

$$\dot{\eta} = \eta \dfrac{1 - exp(-ad_\zeta)}{ad_\zeta} \dot{\zeta}$$

$$
\dot{\zeta} = \dfrac{ad_\zeta}{1 - exp([-ad_\zeta])} \eta^{-1} \dot{\eta} \\
v = \bar{v} + u + w
$$

Where: $\dfrac{1 - exp([-ad_\zeta])}{ad_\zeta} = \sum\limits_{k=0}^{\infty} \dfrac{(-1)^k}{(k+1)!} (ad_\zeta)^k$

If $\dot{\eta} = \eta [\bar{v}] - [v] \eta$ then:

\begin{align}
\eta^{-1} \dot{\eta} &= [\bar{v} - Ad_{\eta^{-1}} (\bar{v} + (u + w))] \\
&= [(1 - Ad_{\eta^{-1}}) \bar{v} - Ad_{\eta^{-1}} (u + w)] \\
&= [(1 - exp(-ad_\zeta)) \bar{v} - exp(-ad_\zeta) (u + w)] \\
\dot{\zeta} &= \frac{ad_\zeta}{1 - exp(-ad_\zeta)} \left(1 - exp(-ad_\zeta) \right) \bar{v} - exp(-ad_\zeta) (u + w)\\
&= -ad_{\bar{v}} \zeta  -\frac{ad_\zeta exp(-ad_\zeta)}{1 - exp(-ad_\zeta)} (u + w)
\end{align}

$$\boxed{\dot{\zeta} = -ad_{\bar{v}} \zeta   -\frac{ad_\zeta exp(-ad_\zeta)}{1 - exp(-ad_\zeta)} (u + w)}$$



Let $\boxed{U = -\frac{ad_\zeta exp(-ad_\zeta)}{1 - exp(-ad_\zeta)}}$

$$\boxed{\dot{\zeta} = -ad_{\bar{v}} \zeta + U (u + w)}$$

If all elements of the Lie algebra may be used for control, then:

Let $A = -ad_{\bar{v}}$, $u = U^{-1} B K\zeta$
$$\boxed{\dot{\zeta} = \left(A + B K \right) \zeta  + U w}$$

This only holds if all elements of the Lie algebra are valid inputs. For instance, the car cannot move sideways, but the differential correction, might require that for the group to follow the linear system dynamics.

In the case that there is a Lie algebra element that is not usuable, the difference can be treated as a disturbance. For instance, consider the matrix $L = diag([1, 0, 1])$.

Let $A = -ad_{\bar{v}}$, $u = L U^{-1} B K\zeta$

If all elements of the Lie algebra may be used for control, then:
$$\boxed{\dot{\zeta} = \left(A + B K \right) \zeta  + (ULU^{-1} - I) BK \zeta + U w}$$


For SE(2), the U matrix is:

$$U = \left[\begin{matrix}- \frac{\theta \sin{\left(\theta \right)}}{2 \left(\cos{\left(\theta \right)} - 1\right)} & \frac{\theta}{2} & - \frac{\theta x \sin{\left(\theta \right)} + \left(1 - \cos{\left(\theta \right)}\right) \left(\theta y - 2 x\right)}{2 \theta \left(1 - \cos{\left(\theta \right)}\right)}\\- \frac{\theta}{2} & - \frac{\theta \sin{\left(\theta \right)}}{2 \left(\cos{\left(\theta \right)} - 1\right)} & - \frac{\theta y \sin{\left(\theta \right)} + \left(1 - \cos{\left(\theta \right)}\right) \left(- \theta x - 2 y\right)}{2 \theta \left(1 - \cos{\left(\theta \right)}\right)}\\0 & 0 & 1\end{matrix}\right]$$

With a Taylor series expansion at $\theta=0$ of:

$$U = \left[\begin{matrix}1 - \frac{\theta^{2}}{12} - \frac{\theta^{4}}{720} + O\left(\theta^{5}\right) & \frac{\theta}{2} & - \frac{y}{2} + \frac{\theta x}{12} + \frac{\theta^{3} x}{720} + O\left(\theta^{5}\right)\\- \frac{\theta}{2} & 1 - \frac{\theta^{2}}{12} - \frac{\theta^{4}}{720} + O\left(\theta^{5}\right) & \frac{x}{2} + \frac{\theta y}{12} + \frac{\theta^{3} y}{720} + O\left(\theta^{5}\right)\\0 & 0 & 1\end{matrix}\right]$$

The invariant set for a system, can therefore be calculated, by finding an upper bound for the singular value of the U matrix over the invariant set, since this depends on the invariant set. An iterative process is necessary to find the invariant set.

1. Guess maximum singular value in invariant set
2. Compute invariant set using LMI by scaling disturbance by maximum singular value
3. Calculate maximum singular value over invariant set
4. Adjust new maximum starting singular value guess 
5. Repeat steps 1-4 until converged to desired tolerance (favor over approximation for safety)

# Simulation

```python
%load_ext autoreload
%autoreload 2
import se2_flow
import numpy as np

freq = [0.1]
res = []
t = 40
# disturbances
w1 = 1
w2 = 1


for theta0 in np.linspace(-np.pi/2, np.pi/2, 2):
    for x0 in np.linspace(-1, 1, 2):
        for y0 in np.linspace(-1, 1, 2):
                for i in freq:
                    r = se2_flow.simulate(40, i, w1, w2, x0, y0, theta0, dist = 'square', sol='', use_approx=False)
                    if theta0 == np.pi/2 and x0 == 1 and y0 == 1 and i == freq[-1]:
                        se2_flow.plot_simulated_corres(r, name='', legend=True, save=True)
                    else:
                        se2_flow.plot_simulated_corres(r, name='', legend=False, save=False)
```
